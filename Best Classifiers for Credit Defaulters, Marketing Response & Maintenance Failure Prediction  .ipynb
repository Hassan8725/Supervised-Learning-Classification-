```python
df1= pd.read_csv("C:/Users/Hassan Ahmed/Desktop/Data Science Cohort-1 FTI/creditriskmodeling.csv")
df2= pd.read_csv("C:/Users/Hassan Ahmed/Desktop/Data Science Cohort-1 FTI/direct.marketing.csv")
df3= pd.read_csv("C:/Users/Hassan Ahmed/Desktop/Data Science Cohort-1 FTI/predictivemaintenance.csv")
display(df1.head())
display(df2.head())
display(df3.head())
```

### Little Preprocessing

```python
df1= df1.dropna(axis=0)
df2=df2.drop(['Name','Email'],axis=1)
df2 = pd.get_dummies(df2,drop_first=True)
```

## X and Y of all three Dataframes

```python
# For Credit Risk Modeling
x1=df1.drop('Default',axis=1)
x1cols=x1.columns
x1_s= skp.StandardScaler().fit(x1).transform(x1)
x1_sdf= pd.DataFrame(skp.StandardScaler().fit(x1).transform(x1))
y1o=df1['Default'].copy()
level_map = {'Yes':1, 'No':0}
y1 = y1o.map(level_map)
# For Direct Marketing
x2=df2.drop('Response_yes',axis=1)
x2cols=x2.columns
x2_s = skp.StandardScaler().fit(x2).transform(x2)
x2_sdf = pd.DataFrame(skp.StandardScaler().fit(x2).transform(x2))
y2=df2['Response_yes'].copy()
# For Predictive Maintenance
x3=df3.drop(['Failure', 'Machine_ID'],axis=1)
x3cols=x3.columns
x3_s=skp.StandardScaler().fit(x3).transform(x3)
x3_sdf=pd.DataFrame(skp.StandardScaler().fit(x3).transform(x3))
y3=df3['Failure'].copy()
level_map1 = {'yes':1, 'no':0}
y3n = y3.map(level_map1)
```

## Gaussian Naive Bayes Algo

```python
# Credit Risk Modeling
accarray1=[] # Create  Accuaracy Array
aucarray1=[] # Area under the Curve Array
precision1=[]
recall1=[]
# Define the split state=99, into 10 folds
kf1=skm.StratifiedKFold(n_splits=4,random_state=99,shuffle=False)
for train_index1, test_index1 in kf1.split(x1_s,y1):
    trainx1, testx1=x1_s[train_index1],x1_s[test_index1]
    trainy1,testy1=y1[train_index1],y1[test_index1]
    clf1=GaussianNB()
    clf1.fit(trainx1,trainy1)
    predictions1=clf1.predict(testx1)
    accarray1=np.append(accarray1,accuracy_score(testy1,predictions1)*100)
    fpr1 , tpr1,_ = roc_curve(testy1, predictions1)
    aucarray1=np.append(aucarray1,auc(fpr1, tpr1)*100)
    precision1=np.append(precision1,precision_score(testy1,predictions1)*100)
    recall1=np.append(recall1,recall_score(testy1,predictions1)*100)
print("Average accuracy of Credit Risk Modeling Data from Gaussian Naive Bayes is: \t", np.mean(accarray1))
print("Average accuracy of Credit Risk Modeling Data Data from Gaussian Naive Bayes is: \t", np.mean(accarray1))
print("Average AUC of Credit Risk Modeling  Data from Gaussian Naive Bayes is: \t",np.mean(aucarray1))
print("Average Precision of Credit Risk Modeling  Data from Gaussian Naive Bayes is: \t",np.mean(precision1))
print("Average Recall of Credit Risk Modeling Data from Gaussian Naive Bayes is: \t",np.mean(recall1))
```

```python
# Direct Marketing
accarray2=[] # Create  Accuaracy Array
aucarray2=[] # Area under the Curve Array
precision2=[]
recall2=[]
# Define the split state=99, into 10 folds
kf2=skm.StratifiedKFold(n_splits=4,random_state=99,shuffle=False)
for train_index2, test_index2 in kf2.split(x2_s,y2):
    trainx2, testx2=x2_s[train_index2],x2_s[test_index2]
    trainy2,testy2=y2[train_index2],y2[test_index2]
    clf2=GaussianNB()
    clf2.fit(trainx2,trainy2)
    predictions2=clf2.predict(testx2)
    accarray2=np.append(accarray2,accuracy_score(testy2,predictions2)*100)
    fpr2 , tpr2,_ = roc_curve(testy2, predictions2)
    aucarray2=np.append(aucarray2,auc(fpr2, tpr2)*100)
    precision2=np.append(precision2,precision_score(testy2,predictions2)*100)
    recall2=np.append(recall2,recall_score(testy2,predictions2)*100)
print("Average accuracy of Direct Marketing Data from Gaussian Naive Bayes is: \t", np.mean(accarray2))
print("Average AUC of Direct Marketing Data from Gaussian Naive Bayes is: \t",np.mean(aucarray2))
print("Average Precision of Direct Marketing Data from Gaussian Naive Bayes is: \t",np.mean(precision2))
print("Average Recall of Direct Marketing Data from Gaussian Naive Bayes is: \t",np.mean(recall2))
```
